manipulating data with mongodb

```python
import pymongo
import json
import datetime
```

```python
client=pymongo.MongoClient("mongodb://localhost:30001")
```

```python
mydb=client["TSLA_TWEETS"]
```

```python
# make a collection for each day
datetime_object = datetime.datetime.strptime('2021-01-01 23:49:43+00:00', '%Y-%m-%d %H:%M:%S+00:00')
datetime_object.strftime('%Y-%m-%d')
```

```python

with open("TSLA_tweets.json", "r") as f:
    for line in f:
        data = json.loads(line)
        date = data["date"]
        date = datetime.datetime.strptime(date, '%Y-%m-%d %H:%M:%S+00:00')
        date = date.strftime('%Y-%m-%d')
        mycol = mydb[date]
        mycol.insert_one(data)  # Pass the dictionary 'data' instead of 'json.loads(line)'
f.close()
```

```python
col=mydb["2021-06-04"]
```

```python
col.find_one()
```

```python
for record in col.find({},{"rawContent":1,"_id":0,"user":1,"likeCount":1,"retweetCount":1,"replyCount":1,"quoteCount":1}):
    print(record["rawContent"],end="\n")
    print("followers:",record["user"]["followersCount"],"likes:",record["likeCount"],"retweets:",record["retweetCount"],"replies:",record["replyCount"],"quotes:",record["quoteCount"], end="\n"+"-"*100+"\n\n",sep="\t")
```

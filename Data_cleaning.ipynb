manipulating data with mongodb

```python
import pymongo
import json
import datetime
```

```python
client=pymongo.MongoClient("mongodb://localhost:30001")
```

```python
mydb=client["TSLA_TWEETS"]
```

```python
# make a collection for each day
datetime_object = datetime.datetime.strptime('2021-01-01 23:49:43+00:00', '%Y-%m-%d %H:%M:%S+00:00')
datetime_object.strftime('%Y-%m-%d')
```

```python

with open("TSLA_tweets.json", "r") as f:
    for line in f:
        data = json.loads(line)
        date = data["date"]
        data["_id"]=data["id"]
        del data["id"]
        date = datetime.datetime.strptime(date, '%Y-%m-%d %H:%M:%S+00:00')
        date = date.strftime('%Y-%m-%d')
        mycol = mydb[date]
        try:
            mycol.insert_one(data)  
        except:
            continue
        # Pass the dictionary 'data' instead of 'json.loads(line)'
f.close()
```

```python
import os
# read for each file in folder GET_OLD_TWEETS
for filename in os.listdir("GET_OLD_TWEETS"):
    with open("GET_OLD_TWEETS/"+filename, "r") as f:
        for line in f:
            data = json.loads(line)
            date = data["date"]
            data["_id"]=data["id"]
            del data["id"]
            date = datetime.datetime.strptime(date, '%Y-%m-%d %H:%M:%S+00:00')
            date = date.strftime('%Y-%m-%d')
            mycol = mydb[date]
            try:
                mycol.insert_one(data)  
            except:
                continue
            # Pass the dictionary 'data' instead of 'json.loads(line)'
    f.close()
```

```python
# drop collections before 2019-01-01
mydb = client["TSLA_TWEETS"]
for col in mydb.list_collection_names():
    if (datetime.datetime.strptime(col,'%Y-%m-%d') < datetime.datetime.strptime("2019-01-01",'%Y-%m-%d')) or (datetime.datetime.strptime(col,'%Y-%m-%d') > datetime.datetime.strptime("2023-11-24",'%Y-%m-%d')):
        mycol = mydb[col]
        mycol.drop()
```

```python
#transform to dataframe
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##create dataframe

df_followerbased = pd.DataFrame(columns=["date","followers","likes","retweets","replies","quotes","RawContent"])
```

```python
df_temp = pd.DataFrame({"date": [], "followers": [], "likes": [], "retweets": [], "replies": [], "quotes": [], "RawContent": []})
for col in mydb.list_collection_names():
    for record in mydb[col].find({}, {"rawContent": 1, "_id": 0, "user": 1, "likeCount": 1, "retweetCount": 1, "replyCount": 1, "quoteCount": 1}):
        df_temp = pd.concat([df_temp, pd.DataFrame({"date": [col], "followers": [record["user"]["followersCount"]], "likes": [record["likeCount"]], "retweets": [record["retweetCount"]], "replies": [record["replyCount"]], "quotes": [record["quoteCount"]], "RawContent": [record["rawContent"]]})])
df_followerbased = pd.concat([df_followerbased, df_temp], ignore_index=True)
```
